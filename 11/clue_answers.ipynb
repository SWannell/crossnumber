```python
int2list = lambda x: [str(x)[i] for i in range(len(str(x)))]
int2intlist = lambda x: [int(str(x)[i]) for i in range(len(str(x)))]
digsum = lambda n: sum([i for i in int2intlist(n)])
reverse2str = lambda n: str(n)[::-1]
reverse2int = lambda n: int(reverse2str(n))
```

- 7D: This number is equal to the sum of the fifth powers of its digits. (5)
- 31D: This number is equal to the sum of the fifth powers of its digits. (5)

```python
d31 = []
for n in range(10000, 100000):
    _ = int2intlist(n)
    fifth_sum = sum([i**5 for i in _])
    if fifth_sum == n:
        d31.append(n)
        print(n)
d7 = list(d31)
```

- 15A: This number is a multiple of each of its digits. (7)
- 29A: This number is a multiple of each of its digits. (7)

```python
# 1000000
for i in range(1000000, 10000000):
    _ = int2intlist(i)
    if 0 in _:
        continue
    rems = set([i % dig for dig in _])
    if rems == set([0]):
        print(i)
```

```python
_ = int2intlist(12)
if 0 in _:
    print("fail")
for dig in _:
    if 12 % dig != 0:
        break
print(12)
```
